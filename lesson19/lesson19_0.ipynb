```python
import pandas_datareader.data as pdr
import yfinance as yf

yf.pdr_override()
all_data = {item:pdr.get_data_yahoo(item) for item in ['2330.TW','2303.TW','2454.TW','2317.TW']}
all_data
```

```python
import pandas as pd
items = {ticker:data['Adj Close']for ticker,data in all_data.items()}
stock_price = pd.DataFrame(items)
stock_price.columns = ['台積電','聯電','聯發科','鴻海']
stock_price
```

```python
stock_price.info()
#發現許多不用的資料需要整理
```

```python
df1=stock_price.dropna()
df1.info()
df1
```

```python
df1.query('台積電>=500')
```

```python
#date (型別為DatetimeIndex:)是個索引
df1.loc['2023-01-04'] 
```

```python
#型別是datetime 的好處:直接月份抓資料
df1.loc['2023-01'] 
```

```python
#型別是datetime 的好處:直接用年份抓資料
df1.loc['2023'] 
```

```python
stock2023 = df1.loc['2023-03-05':'2023-05-05']
stock2023
```

```python
pctchange = df1.loc['2023'].pct_change().dropna()
pctchange
#當前元素與先前元素之間的分數變化。
```

```python
stock2023_1 = (stock2023 - stock2023.shift(1)).dropna()
stock2023_1
```
