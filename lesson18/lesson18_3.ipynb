```python
import numpy as np
import pandas as pd

scores=pd.DataFrame(np.random.randint(50,101,size=(50,5)),
                    columns=['國文', '英文', '數學', '地理', '歷史'],
                    index=range(1,51))
scores
```

```python
#lambda 匿名函式，是在電腦的編碼過程中，不需要定義函式名稱的一種函式。目前這種函式已經普遍的存在多種程式語言中。在Python中所謂Lambda函式，也就是匿名函式。
#函式本身可以接收多個參數並且返回ㄧ個表達式的值。其中func是指定的名稱，也可以沒有。
#沿 DataFrame 的軸應用函數。

scores.apply(lambda item:print(item))
#DataFrame.apply(func, axis=0, raw=False, result_type=None, args=()....
#apply內要放一個funw它就會沿著 DataFrame 的軸去應用函數,並用serise的方式回傳給我們。
```

```python
def abc(s):
    return s.max()
scores.apply(abc)

#最大只有一個值
```

```python
def abc(s):
    return pd.Series([s.max(),s.min(),s.mean()],index=['最高分','最低分','平均'])
scores.apply(abc)

#每個值分別輸出後都是series的形式，組合起來後會變成Dataframe的形式。
```

```python
scores.index = scores.index.map(lambda i:f'{i}號')
scores.index = scores.index.str.replace(r'號+', '號', regex=True)
scores
```

```python
def abc1(item):
    return pd.Series([item.sum(),item.mean()],index=['總分','平均'])
scores.apply(abc1,axis=1)
```

```python
one_student = scores.iloc[2]
less_student = one_student[one_student < 60]
less_student
```

```python
#重新定義function
def abc1(item):
    less_student=item[item<60]
    less_count=less_student.count()
    less_str = ",".join(less_student.index.values.tolist())
    
    return pd.Series([item.sum(),item.mean(),less_count,less_str],index=['總分','平均','不及格科目數','不及格科目'])
scores.apply(abc1,axis=1)
#利用apply 針對每一列進行分析並得到總分、平均、不及格科目數,不及格科目
```
