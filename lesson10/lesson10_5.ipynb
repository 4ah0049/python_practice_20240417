```python
from csv import DictReader
with open ('個股日成交資訊.csv',encoding='utf-8',newline='') as file:
     reader:DictReader=DictReader(file)
     stock_list:list=list(reader)
     #print(type(stock_list))

for stock_lists in stock_list:
     print('{:<8}'.format(stock_lists['證券代號']),'{:<10}'.format(stock_lists['證券名稱']),'{:>}'.format(stock_lists['成交股數']))
```
