```python
t = np.arange(0,5,0.1) #傳回給定間隔內均勻分佈的值。
y1 = np.sin(2 * np.pi * t)
y2 = np.cos(2 * np.pi * t)

fig = plt.figure(figsize=(8,15))
ax1 = fig.add_subplot(2,1,1) #兩列一欄的第一個子圖
ax2 = fig.add_subplot(2,1,2) #兩列一欄的第二個子圖

ax1.plot(t,y1,"b-.")
ax2.plot(t,y2,"r-.")
plt.show()
```

```python
#一維的操作跟list相同
arr = np.arange(10)
#索引編號
arr[1] #subscript[index]
arr[:] #subscirpt[:] slice
arr[:5] #subscirpt[:] slice
arr[-5:] #subscirpt[:] slice(切割)
```
