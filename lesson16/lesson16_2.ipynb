```python
#list comprehension->快速建立list
[f'{num}號' for num in range(1,11)]
```

```python
import numpy as np
import pandas as pd
array_2d = np.random.randint(50,100,size=(50,5))
df1 = pd.DataFrame(array_2d,
             columns=['國文','英文','數學','地理','社會'],
             index = [f'{num}號' for num in range(1,51)] #建立出的list 當作index 1號~50號
             )
df1
```

```python
#從Dataframe取出ndarry
df1.values  #傳回 DataFrame 的 Numpy 表示形式。
```

```python
#dataFrame[欄位名稱]     []中括號就是-->subscript
type(df1['國文']) 

#dataframe是由多個以上的Series組成的,而Series只有index沒有欄位。
```

```python
df1['國文'],df1['英文']
```

```python
#Series的選取,使用subscript
#Series[數值的index]
df1['國文'][0] 

```

```python
#Series[數值的slice]
df1['國文'][:5]

```

```python
#Series[字串的index]
df1['國文']['1號']
```

```python
#Series[[字串,字串,字串]]
df1['國文'][['1號','2號','3號']]
```

```python
#Series[[list comprehenis]]
df1['國文'][[f'{num}號' for num in range(1,11)]]
```

```python
#dataFrame的subscripts
#dataFrame[欄位名稱]
df1['國文']
```

```python
#dataFrame的subscripts
#dataFrame[[欄位名稱,欄位名稱]]
df1[['國文','英文']] #兩個series[[]]-->變成Dataframe 
```
