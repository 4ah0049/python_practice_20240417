```python
student1 = {'chinese':78, 'english':80, 'math':92, 'history':75, 'discover':85}
student2 = {'chinese':92, 'english':60, 'math':85, 'history':62, 'discover':73}
keys = student1.keys() #可以轉成list
#print(keys)
subject:list[str] = list(keys)
values = student1.values()
values1 = list(values)
values2 = list(student2.values())

#繪圖
import matplotlib.pyplot as plt

figure = plt.figure(figsize=(8,5),dpi=250,facecolor='#EEA9A9') #日本色標
axes = figure.add_subplot(1, 1, 1)

axes.plot(subject,values1,'bo--') #[color][marker][line]決定線的相關細節
axes.plot(subject,values2,'ro-.')

plt.show()
```
